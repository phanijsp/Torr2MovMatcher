```python
import torch
import torch.version
from transformers import BertTokenizer, BertModel
from sklearn.metrics.pairwise import cosine_similarity
import random

# Check if GPU is available
device = torch.device("cuda")
print(torch.cuda.is_available())
print(torch.version.cuda)

# Load pre-trained BERT model and tokenizer
model_name = 'bert-base-uncased'
tokenizer = BertTokenizer.from_pretrained(model_name)
model = BertModel.from_pretrained(model_name).to(device)

# Input sentences
sentence1 = "I like coding with Python."

# Generate 20 random sentences for comparison
# random_sentences = [
#     "Python is my favorite programming language.",
#     "Machine learning is an exciting field.",
#     "Artificial intelligence has many applications.",
#     # Add more random sentences here
# ]

# Tokenize and encode the sentences
inputs = tokenizer([sentence1] + random_sentences, return_tensors='pt', padding=True, truncation=True).to(device)

# Get BERT embeddings for the sentences
with torch.no_grad():
    outputs = model(**inputs)

# Extract BERT embeddings for the first sentence
embeddings1 = outputs.last_hidden_state[0][0].cpu().numpy()

# Extract BERT embeddings for the random sentences
embeddings_random = [outputs.last_hidden_state[i][0].cpu().numpy() for i in range(1, len(random_sentences) + 1)]

# Calculate cosine similarity between the first sentence and random sentences
similarity_scores = cosine_similarity(embeddings1.reshape(1, -1), embeddings_random)

# Print cosine similarity scores
print("Cosine Similarity Scores:")
for i, score in enumerate(similarity_scores[0]):
    print(f"Sentence {i+1}: {score}")
```
